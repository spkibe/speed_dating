### Using function to make the code Dry, utilize func for the melting process

```python
def reshape_dataframe(df, id_vars_columns, selected_columns, new_value_name):
    """
    Reshapes the given DataFrame from wide format to long format based on specified columns.

    Parameters:
    df (DataFrame): The input DataFrame to be reshaped.
    id_vars_columns (list): List of columns to keep unchanged (id_vars).
    selected_columns (list): List of columns to melt down (value_vars).
    new_value_name (str): The new name for the melted 'value' column.

    Returns:
    DataFrame: The reshaped DataFrame with the specified 'value' column renamed.
    """
    # Reshape DataFrame from wide format to long format
    df_long = pd.melt(df, id_vars=id_vars_columns, value_vars=selected_columns)

    # Rename the resulting 'value' column
    df_long.rename(columns={'value': new_value_name}, inplace=True)

    return df_long

# Example usage
id_vars_columns = ['Event', 'Timestamp', 'Email address', 'Your name', 'Mobile number',
                   'Date_badge', 'Gender', 'consent_to_share_contact_with_a_march',
                   'consent_to_share_pictures']

#new_value_name = 'person_met'

df_male_meets = reshape_dataframe(male_df, id_vars_columns, selected_columns, 'person_met')
df_male_meets.head()
```

```python
def process_crosstab(event_df):
    """
    Processes the given DataFrame to create a crosstab of male and female decisions.

    Parameters:
    event_df (DataFrame): The input DataFrame containing 'Date_Decision_y' and 'Date_Decision_x'.

    Returns:
    DataFrame: A crosstab DataFrame with renamed index and columns, reordered to include all combinations.
    """
    # Create the crosstab
    crosstab_result = pd.crosstab(event_df['Date_Decision_y'], event_df['Date_Decision_x'])

    # Rename index and columns
    crosstab_result.index.name = 'Female Decision'
    crosstab_result.columns.name = 'Male Decision'

    # Reorder and ensure all combinations are present
    crosstab_result = crosstab_result.reindex(index=['Like', 'Friendship', 'Neither'], columns=['Like', 'Friendship', 'Neither'])

    return crosstab_result

# try
result = process_crosstab(oct_2019)
print(result)
```

```python
def get_mutual_likes(event_df):
    """
    Filters the DataFrame for mutual likes and returns a DataFrame with specific columns.

    Parameters:
    event_df (DataFrame): The input DataFrame containing the event data.

    Returns:
    DataFrame: A DataFrame containing the mutual likes with specified columns only.
    """
    # Filter for mutual likes
    mutual_likes = event_df[(event_df['Date_Decision_x'] == 'Like') & (event_df['Date_Decision_y'] == 'Like')]

    # Select specific columns
    selected_columns = ['Event_x', 'Email address_x', 'Your name_x', 'Date_Decision_x',
                        'Event_y', 'Email address_y', 'Your name_y', 'Date_Decision_y']

    # Return the filtered DataFrame with the selected columns
    return mutual_likes[selected_columns]

# Example usage
result = get_mutual_likes(may_22)
result
```

```python
import os
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Set up email details from environment variables
sender_email = os.getenv("SENDER_EMAIL")
password = os.getenv("EMAIL_PASSWORD")
smtp_server = os.getenv("SMTP_SERVER")
smtp_port = int(os.getenv("SMTP_PORT"))

# Function to create the email message
def create_message(sender_email, receiver_email, subject, body, html_content=None):
    message = MIMEMultipart("alternative")
    message["From"] = sender_email
    message["To"] = receiver_email
    message["Subject"] = subject

    # Attach the plain text part to the message
    text_part = MIMEText(body, "plain")
    message.attach(text_part)

    # Optionally attach the HTML part if provided
    if html_content:
        html_part = MIMEText(html_content, "html")
        message.attach(html_part)

    return message

# Function to send email
def send_email(to_email, subject, body, html_content=None):
    try:
        # Establish a connection to the SMTP server
        with smtplib.SMTP(smtp_server, smtp_port) as server:
            server.starttls()  # Secure the connection
            server.login(sender_email, password)  # Log in to the SMTP server

            # Create the message
            message = create_message(sender_email, to_email, subject, body, html_content)

            # Send the email
            server.sendmail(sender_email, to_email, message.as_string())
            print(f"Email sent successfully to {to_email}")
    except Exception as e:
        print(f"Failed to send email to {to_email}: {e}")

# Function to notify participants based on the decision matrix
def notify_participants(df):
    for index, row in df.iterrows():
        name_a = row['Your name_x']
        email_a = row['Email address_x']
        decision_a = row['Date_Decision_x']
        
        name_b = row['Your name_y']
        email_b = row['Email address_y']
        decision_b = row['Date_Decision_y']
        
        if decision_a == 'Like' and decision_b == 'Like':
            subject = "It's a Match!"
            body = f"Hi {name_a} and {name_b},\n\nYou both liked each other! Feel free to get in touch.\n\nBest,\nYour Dating App"
            send_email(email_a, subject, body)
            send_email(email_b, subject, body)
        
        elif (decision_a in ['Like', 'Friendship']) and (decision_b in ['Like', 'Friendship']):
            subject_a = "New Connection"
            body_a = f"Hi {name_a},\n\n{name_b} has shown interest in {decision_b}. You can reach out to them!\n\nBest,\nYour Dating App"
            send_email(email_a, subject_a, body_a)
            
            subject_b = "New Connection"
            body_b = f"Hi {name_b},\n\n{name_a} has shown interest in {decision_a}. You can reach out to them!\n\nBest,\nYour Dating App"
            send_email(email_b, subject_b, body_b)

```
